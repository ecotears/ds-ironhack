```python
driver_data = {}
trip_data = []

#Adding drivers to the dictionary
driver_data[1] = {"full_name": "Mona Lisa", "contact_number": "333-444-555", "total_trips": 0}
driver_data[2] = {"full_name": "Mariano Eusebio", "contact_number": "111-111-111", "total_trips": 0}
driver_data[3] = {"full_name": "Lucia Lopez", "contact_number": "000-002-111", "total_trips": 0}

#Recording taxi trips
trip_data.append({"trip_id": 111, "driver_id": 1, "customer_name": "Pepito Perez", "distance": 20.5, "fare": 11.2})
trip_data.append({"trip_id": 222, "driver_id": 3, "customer_name": "Lisa Lu", "distance": 1.2, "fare": 2})
trip_data.append({"trip_id": 333, "driver_id": 2, "customer_name": "Ana Garcia", "distance": 100, "fare": 60})
trip_data.append({"trip_id": 444, "driver_id": 3, "customer_name": "Lola Garzon", "distance": 50.2, "fare": 33.3})
```

```python
# Calculating and updating the total number of trips
for i in trip_data:
    var_driver_id = i['driver_id'] 
    if var_driver_id in driver_data.keys():
        driver_data[var_driver_id]['total_trips'] =  driver_data[var_driver_id]['total_trips'] + 1

print(driver_data)
```

```python
# Calculate total revenue generated by the taxi company

revenue = 0
for i in trip_data:
    revenue += i["fare"]

print(f"El total revenue es de {revenue}")
                    
```

```python
#Find the driver with the highest number of trips.

maximum_trips = 0

for i,j in driver_data.items():
    if j['total_trips'] > maximum_trips:
        maximum_trips = j['total_trips']
    else:
        print(f"El que más trips tiene es {j['full_name']}")
```

```python
#Find the driver with the highest total revenue

#Encontrar primero al taxista
#Luego total de distancia * total de fare
#Luego imprimir el que más gane

max_revenue = 0
max_driver_id = 0

for driver_id, revenue in driver_data.items():
   temp_rev = sum(trips["fare"] for trips in trip_data if trips['driver_id'] == driver_id)
   if temp_rev > max_revenue:
       max_revenue = temp_rev
       max_driver_id = driver_id


print (f"El driver con número {max_driver_id} ha generado la mayor cantidad de revenue, siendo {max_revenue}")
```

```python
# List all the trips conducted by a specific driver

driverid_toprint = 3

for driver_id in trip_data:
    if driverid_toprint == driver_id['driver_id']:
        print(f"""Trip
        {driver_id['trip_id'], driver_id['driver_id'], driver_id['customer_name'],driver_id['distance'],driver_id['fare']}""")
```
